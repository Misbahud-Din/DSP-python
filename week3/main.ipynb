## Importing Numpy and Matplotlib Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Calculating Values for Discrete Signal

```python
n1 = np.arange(-10, 10)
y1 = np.cos((3 / 2) * np.pi * n1)
y2 = np.cos((7 / 2) * np.pi * n1)
y3 = np.cos((1 / 2) * np.pi * n1)
```

## Plotting different discrete signals in one plot

```python
plt.stem(n1, y1)
plt.stem(n1, y2)
plt.stem(n1, y3)
plt.figure()
# All discrete signals plotted have same time period
```

## Calculating Values for Continuous Signal

```python
t1 = np.arange(-5, 5, 0.1)
y1 = np.cos((3 / 2) * np.pi * t1)
y2 = np.cos((7 / 2) * np.pi * t1)
y3 = np.cos((1 / 2) * np.pi * t1)
```

## Plotting different continous signals in one plot

```python
plt.plot(t1, y1)
plt.plot(t1, y2)
plt.plot(t1, y3)
plt.figure()
# Continuous time signal having different time periods
```
